```python
import numpy as np
import pandas as pd
from sklearn import linear_model
```

```python
data = pd.read_csv('/content/drive/MyDrive/Machine Learning hub/Machine Learning specialization cousera/Supervised Machine Learning/Real_Estate.csv')
data.head()
```

```python
#lets select a major feartures
data_updated = data.loc[:,['House age','Distance to the nearest MRT station','Number of convenience stores','Latitude','Longitude','House price of unit area']]
data_updated.head()
```

```python
#to splite into dependent and independent varibales

x = data_updated.drop(['House price of unit area'], axis  = 1)
y = data_updated['House price of unit area']
y
```

```python
#to splite the data into train and test dataset
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size = 0.2 , random_state = 42)
```

```python
#lets standardarize data
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
x_train_scaled = scalar.fit_transform(x_train)
x_test_scaled = scalar.transform(x_test)
```

# fit_transform() vs transform()
Purpose: This method is used to apply the previously computed transformation (using fit()) to a new dataset (like the test set).
When to Use: This is used on the test dataset to apply the same transformation that was applied to the training data. It ensures that the test data is scaled in the same way as the training data, using the statistics computed from the training set.
What Happens: The transform() method applies the transformation using the parameters (like mean and standard deviation) that were learned from the training data via fit_transform(). It does not recompute these parameters.
- Consistency Between Training and Test Data:

```python
#check for null values
data_updated.isnull().sum() #no null values
```

```python
#lets train our liner model

lr = linear_model.LinearRegression()
model = lr.fit(x_train_scaled,y_train)
predictions = model.predict(x_test_scaled)
predictions
```

```python
#on test data
model.score(x_test_scaled,y_test)
```

```python
#performence on train data
model.score(x_train_scaled,y_train)
```

## lets apply  Lasso, Ridge Regression


```python
from sklearn.linear_model import Lasso, Ridge

# Lasso
lasso = Lasso()
lasso.fit(x_train_scaled, y_train)


# Ridge
ridge = Ridge()
ridge.fit(x_train_scaled, y_train)


```

```python
#after regularization let predict score

ridge.predict(x_test_scaled)
lasso.predict(x_test_scaled)

print(f'ridge_regularization score on test data {ridge.score(x_test_scaled,y_test)}' )
print(f'lasso_regularization score on test data {lasso.score(x_test_scaled,y_test)}' )
#train data
print("\t")
print(f'ridge_regularization score on train data {ridge.score(x_train_scaled,y_train)}' )
print(f'lasso_regularization score on train data {lasso.score(x_train_scaled,y_train)}' )
```
